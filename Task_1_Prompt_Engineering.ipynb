## Task 1: Replace Hardcoded Prompts (10%)

**Goal:** Convert an f-string based function into a reusable LangChain `PromptTemplate`.

```python
# -------------------------------------------------------
# TASK 1: Replace Hardcoded Prompts
# Original hardcoded version (DO NOT USE — for reference only):
#   def explain_topic(topic):
#       return f"Explain {topic} in simple terms for beginners"
# -------------------------------------------------------

# Step 1: Define a reusable PromptTemplate — no f-strings, no hardcoding
explain_template = PromptTemplate(
    input_variables=["topic"],
    template="Explain {topic} in simple terms for beginners"
)

# Step 2: Use the template to generate prompts dynamically
prompt_ml = explain_template.format(topic="Machine Learning")
prompt_ai = explain_template.format(topic="Artificial Intelligence")
prompt_dl = explain_template.format(topic="Deep Learning")

# Step 3: Print results
print("=== Task 1 Output ===")
print(prompt_ml)
print(prompt_ai)
print(prompt_dl)
```

## Task 2: Multi-Input Prompt System (15%)

**Goal:** Build a template that accepts multiple inputs — `topic`, `audience`, and `tone`.

```python
# -------------------------------------------------------
# TASK 2: Multi-Input Prompt System
# Template: Explain {topic} for {audience} in a {tone} tone
# -------------------------------------------------------

# Step 1: Define a PromptTemplate with multiple input variables
multi_input_template = PromptTemplate(
    input_variables=["topic", "audience", "tone"],
    template="Explain {topic} for {audience} in a {tone} tone"
)

# Step 2: Define test cases as a list of dictionaries — no hardcoding inside the template
test_cases = [
    {"topic": "AI",            "audience": "beginners", "tone": "friendly"},
    {"topic": "Python",        "audience": "kids",       "tone": "fun"},
    {"topic": "Deep Learning", "audience": "engineers",  "tone": "technical"},
]

# Step 3: Generate prompts for each test case
print("=== Task 2 Output ===")
for i, case in enumerate(test_cases, start=1):
    prompt = multi_input_template.format(**case)
    print(f"Test Case {i}: {prompt}")
```

## Task 3: Prompt Variations Engine (15%)

**Goal:** Create 3 distinct `PromptTemplate`s — Teaching, Interview, and Storytelling.

```python
# -------------------------------------------------------
# TASK 3: Prompt Variations Engine
# Three separate templates for three different prompt styles
# -------------------------------------------------------

# Template 1: Teaching style
teaching_template = PromptTemplate(
    input_variables=["topic"],
    template="Explain {topic} clearly step by step"
)

# Template 2: Interview style
interview_template = PromptTemplate(
    input_variables=["topic"],
    template="Ask 3 questions about {topic}"
)

# Template 3: Storytelling style
storytelling_template = PromptTemplate(
    input_variables=["topic"],
    template="Explain {topic} as a story"
)

# Organise all templates in a dictionary for clean, modular access
variation_templates = {
    "Teaching":     teaching_template,
    "Interview":    interview_template,
    "Storytelling": storytelling_template,
}

# Run all three templates for the given topic
topic = "Machine Learning"

print("=== Task 3 Output ===")
for style_name, template in variation_templates.items():
    generated = template.format(topic=topic)
    print(f"{style_name} Style  -->  {generated}")
```

## Task 4: ChatPromptTemplate System (15%)

**Goal:** Build a role-based chat prompt system using `ChatPromptTemplate` with System and User messages.

```python
# -------------------------------------------------------
# TASK 4: ChatPromptTemplate System
# Roles: teacher | interviewer | motivator
# System message defines behavior; User message carries dynamic input
# -------------------------------------------------------

# Step 1: Define role-specific system message descriptions
role_descriptions = {
    "teacher":     "You are a knowledgeable and patient teacher. Explain concepts clearly and simply.",
    "interviewer": "You are a professional technical interviewer. Ask insightful questions about the topic.",
    "motivator":   "You are an enthusiastic motivator. Inspire and encourage the learner about the topic.",
}


def build_chat_prompt(role, topic):
    """
    Builds a ChatPromptTemplate dynamically based on the role and topic.

    Args:
        role  (str): One of 'teacher', 'interviewer', 'motivator'
        topic (str): The subject to discuss

    Returns:
        list: Formatted chat messages ready to send to an LLM
    """
    # Retrieve the system description for the requested role
    system_description = role_descriptions.get(role, "You are a helpful assistant.")

    # Build the ChatPromptTemplate using system + human message pair
    chat_template = ChatPromptTemplate.from_messages([
        ("system", "{system_message}"),
        ("human",  "Please help me understand: {topic}"),
    ])

    # Format and return the messages
    return chat_template.format_messages(
        system_message=system_description,
        topic=topic
    )


# Test with the provided example
role  = "teacher"
topic = "Neural Networks"

messages = build_chat_prompt(role=role, topic=topic)

print("=== Task 4 Output ===")
print(f"Role  : {role}")
print(f"Topic : {topic}\n")
for msg in messages:
    print(f"[{msg.type.upper()}]  {msg.content}")

# Demo all three roles
print("\n--- All Roles Demo ---")
for r in ["teacher", "interviewer", "motivator"]:
    msgs = build_chat_prompt(role=r, topic="Neural Networks")
    print(f"\nRole: {r.upper()}")
    for m in msgs:
        print(f"  [{m.type.upper()}] {m.content}")
```

## Task 5: Input Validation Layer (10%)

**Goal:** Validate `audience` and `tone` inputs before passing them to a template. Raise error or use defaults if invalid.

```python
# -------------------------------------------------------
# TASK 5: Input Validation Layer
# Valid audience: beginner, intermediate, expert
# Valid tone    : formal, casual, fun
# -------------------------------------------------------

# Define allowed values as constants
VALID_AUDIENCES  = ["beginner", "intermediate", "expert"]
VALID_TONES      = ["formal", "casual", "fun"]
DEFAULT_AUDIENCE = "beginner"
DEFAULT_TONE     = "casual"


def validate_inputs(audience, tone, use_defaults=True):
    """
    Validates audience and tone inputs against allowed values.

    Args:
        audience     (str):  Target audience
        tone         (str):  Tone of the prompt
        use_defaults (bool): If True, replace invalid with defaults;
                             If False, raise ValueError
    Returns:
        tuple: (validated_audience, validated_tone)
    """
    audience = audience.strip().lower()
    tone     = tone.strip().lower()

    # Validate audience
    if audience not in VALID_AUDIENCES:
        if use_defaults:
            print(f"[WARNING] Invalid audience '{audience}'. Defaulting to '{DEFAULT_AUDIENCE}'.")
            audience = DEFAULT_AUDIENCE
        else:
            raise ValueError(f"Invalid audience '{audience}'. Must be one of: {VALID_AUDIENCES}")

    # Validate tone
    if tone not in VALID_TONES:
        if use_defaults:
            print(f"[WARNING] Invalid tone '{tone}'. Defaulting to '{DEFAULT_TONE}'.")
            tone = DEFAULT_TONE
        else:
            raise ValueError(f"Invalid tone '{tone}'. Must be one of: {VALID_TONES}")

    return audience, tone


print("=== Task 5 Output ===")

a, t = validate_inputs("expert", "formal")
print(f"Case 1 (valid)        -->  audience='{a}', tone='{t}'")

a, t = validate_inputs("student", "fun")
print(f"Case 2 (bad audience) -->  audience='{a}', tone='{t}'")

a, t = validate_inputs("intermediate", "aggressive")
print(f"Case 3 (bad tone)     -->  audience='{a}', tone='{t}'")

print("\nCase 4 (raise error on bad inputs):")
try:
    validate_inputs("child", "robotic", use_defaults=False)
except ValueError as e:
    print(f"[ERROR] {e}")
```

## Task 6: Prompt Generator App (15%)

**Goal:** Create `generate_prompt()` combining validation + template selection + formatting.

```python
# -------------------------------------------------------
# TASK 6: Prompt Generator App
# -------------------------------------------------------

# Style-specific PromptTemplates stored in a dictionary
style_templates = {
    "teaching": PromptTemplate(
        input_variables=["topic", "audience", "tone"],
        template="Explain {topic} for {audience} in a {tone} teaching style, step by step"
    ),
    "interview": PromptTemplate(
        input_variables=["topic", "audience", "tone"],
        template="Ask 3 {tone} interview questions about {topic} suitable for a {audience}"
    ),
    "storytelling": PromptTemplate(
        input_variables=["topic", "audience", "tone"],
        template="Explain {topic} for {audience} in a {tone} storytelling style"
    ),
}

VALID_STYLES = list(style_templates.keys())


def generate_prompt(topic, audience, tone, style):
    """
    Full prompt generation pipeline:
      1. Validate audience and tone
      2. Select correct style template
      3. Format and return the final prompt
    """
    # Step 1 — Validate inputs
    audience, tone = validate_inputs(audience, tone, use_defaults=True)

    # Step 2 — Validate style
    style = style.strip().lower()
    if style not in VALID_STYLES:
        print(f"[WARNING] Invalid style '{style}'. Defaulting to 'teaching'.")
        style = "teaching"

    # Step 3 — Select template and format
    return style_templates[style].format(topic=topic, audience=audience, tone=tone)


print("=== Task 6 Output ===")

examples = [
    ("Neural Networks", "beginner",     "fun",    "storytelling"),
    ("Python",          "intermediate", "casual", "teaching"),
    ("Deep Learning",   "expert",       "formal", "interview"),
    ("AI Ethics",       "beginner",     "fun",    "teaching"),
]

for topic, audience, tone, style in examples:
    result = generate_prompt(topic, audience, tone, style)
    print(f"\n--> {result}")
```

## Task 7: Template Reusability Test (10%)

**Goal:** ONE template, 5 different inputs → same structure, different outputs.

```python
# -------------------------------------------------------
# TASK 7: Template Reusability Test
# -------------------------------------------------------

# Single reusable template
reusable_template = PromptTemplate(
    input_variables=["topic", "audience", "tone"],
    template="Explain {topic} for {audience} in a {tone} tone"
)

# 5 different input sets
input_sets = [
    {"topic": "Machine Learning",   "audience": "beginner",     "tone": "fun"},
    {"topic": "Neural Networks",    "audience": "intermediate",  "tone": "casual"},
    {"topic": "Data Science",       "audience": "expert",        "tone": "formal"},
    {"topic": "Python Programming", "audience": "beginner",      "tone": "casual"},
    {"topic": "Cloud Computing",    "audience": "intermediate",  "tone": "fun"},
]

print("=== Task 7 Output ===")
print(f"Template: '{reusable_template.template}'")
print("-" * 65)

for idx, inputs in enumerate(input_sets, start=1):
    output = reusable_template.format(**inputs)
    print(f"Run {idx}: {output}")

print("-" * 65)
print("Observation: Same structure, 5 unique outputs — reusability verified!")
```

## Full Pipeline Demo

`User Input → Validation → Prompt Template → Dynamic Prompt Generation → Output`

```python
# -------------------------------------------------------
# FULL PIPELINE DEMO — all tasks tied together
# -------------------------------------------------------

def run_full_pipeline(topic, audience, tone, style, role=None):
    print("\n" + "=" * 65)
    print("FULL PIPELINE RUN")
    print("=" * 65)
    print(f"  Input  -->  topic='{topic}', audience='{audience}', tone='{tone}', style='{style}'")

    validated_audience, validated_tone = validate_inputs(audience, tone)
    print(f"  After Validation  -->  audience='{validated_audience}', tone='{validated_tone}'")

    final_prompt = generate_prompt(topic, validated_audience, validated_tone, style)
    print(f"  Generated Prompt  -->  {final_prompt}")

    if role:
        chat_msgs = build_chat_prompt(role=role, topic=topic)
        print("  Chat Messages:")
        for msg in chat_msgs:
            print(f"    [{msg.type.upper()}] {msg.content}")

    print("=" * 65)


# Run 1: Valid inputs
run_full_pipeline(
    topic="Neural Networks", audience="beginner",
    tone="fun", style="storytelling", role="teacher"
)

# Run 2: Invalid inputs — validation catches and defaults
run_full_pipeline(
    topic="Deep Learning", audience="engineers",
    tone="technical", style="interview", role="interviewer"
)
```
